```python
# ✅ 0. Config 클래스 및 라이브러리 임포트
import json
import os
import urllib.request
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, models, callbacks

# ----------------------------------------------------------------
# 하이퍼파라미터 및 주요 설정값을 관리하는 클래스
# ----------------------------------------------------------------
class Config:
    VOCAB_SIZE = 10000  # 단어 집합의 크기
    MAX_LEN = 32        # 문장의 최대 길이
    EMBEDDING_DIM = 128 # 임베딩 벡터의 차원
    EPOCHS = 20         # 최대 학습 횟수 (EarlyStopping에 의해 조절됨)
    BATCH_SIZE = 64     # 배치 크기
    MODEL_SAVE_PATH = '/content/rgbcmyw_emotion_model.h5'
    TOKENIZER_SAVE_PATH = '/content/rgbcmyw_tokenizer.json'

# ✅ 1. 데이터 다운로드 및 불러오기
file_name = 'aihub_coupus.csv'
file_path = f'/content/{file_name}'
url = 'https://github.com/ohgzone/file1/raw/main/aihub_coupus.csv'

if not os.path.exists(file_path):
    print(f"'{file_name}'을 다운로드합니다...")
    urllib.request.urlretrieve(url, file_path)
    print("다운로드 완료!")
else:
    print(f"'{file_name}'이 이미 존재합니다.")

df = pd.read_csv(file_path)

# ✅ 2. 전처리 및 RGB/CMY/W 라벨 매핑
# (수정) 실제 컬럼명('감정', '문장')을 코드와 호환되도록 변경
df.rename(columns={'문장': 'utterance', '감정': 'emotion'}, inplace=True)

df = df[['utterance', 'emotion']].dropna()
emotion_map = {
    '분노': 'r', '슬픔': 'b', '기쁨': 'y', '행복': 'y',
    '공포': 'c', '혐오': 'm', '놀람': 'w',
    '중립': 'g', 'neutral': 'g'
}
df['label'] = df['emotion'].map(emotion_map)
df = df.dropna(subset=['label'])

# 샘플 데이터 추가
new_data = [
    {'utterance': '너무 무서워서 소름이 돋았어', 'emotion': '공포', 'label': 'c'},
    {'utterance': '그건 정말 끔찍하고 혐오스러워', 'emotion': '혐오', 'label': 'm'},
    {'utterance': '와! 깜짝 놀랐잖아!', 'emotion': '놀람', 'label': 'w'},
    {'utterance': '그냥 그래', 'emotion': '중립', 'label': 'g'}
]
df = pd.concat([df, pd.DataFrame(new_data)], ignore_index=True)


# ✅ 3. 텍스트 토크나이징 및 시퀀싱
tokenizer = Tokenizer(num_words=Config.VOCAB_SIZE, oov_token='<OOV>')
tokenizer.fit_on_texts(df['utterance'])
sequences = tokenizer.texts_to_sequences(df['utterance'])
x = pad_sequences(sequences, maxlen=Config.MAX_LEN)

# ✅ 4. 토크나이저 저장
tokenizer_json = tokenizer.to_json()
with open(Config.TOKENIZER_SAVE_PATH, 'w', encoding='utf-8') as f:
    f.write(json.dumps(tokenizer_json, ensure_ascii=False))

# ✅ 5. 라벨 인코딩
le = LabelEncoder()
all_labels = list(set(emotion_map.values()))
le.fit(all_labels)
y = le.transform(df['label'])
num_classes = len(le.classes_)
print("\n라벨 매핑:", dict(zip(le.classes_, le.transform(le.classes_))))

# ✅ 6. 데이터 분할
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# ✅ 7. 모델 정의 (다중 필터 1D CNN)
def build_model():
    input_layer = tf.keras.Input(shape=(Config.MAX_LEN,))
    embedding = layers.Embedding(Config.VOCAB_SIZE, Config.EMBEDDING_DIM)(input_layer)
    dropout_emb = layers.Dropout(0.3)(embedding)

    conv_blocks = []
    filter_sizes = [3, 4, 5]
    for size in filter_sizes:
        conv = layers.Conv1D(128, size, activation='relu', padding='valid')(dropout_emb)
        pool = layers.GlobalMaxPooling1D()(conv)
        conv_blocks.append(pool)

    x = layers.Concatenate()(conv_blocks)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    output = layers.Dense(num_classes, activation='softmax')(x)

    model = models.Model(inputs=input_layer, outputs=output)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

model = build_model()
model.summary()

# ✅ 8. 모델 학습
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=Config.EPOCHS,
    batch_size=Config.BATCH_SIZE,
    callbacks=[early_stopping]
)

# ✅ 9. 모델 평가
print("\n===== 최종 평가 결과 =====")
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"테스트 정확도: {test_acc:.4f}")

# ✅ 10. 학습 과정 시각화
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ✅ 11. 모델 및 토크나이저 저장
model.save(Config.MODEL_SAVE_PATH)
print(f"✅ 모델 저장 완료: {Config.MODEL_SAVE_PATH}")
print(f"✅ 토크나이저 저장 완료: {Config.TOKENIZER_SAVE_PATH}")

# ✅ 12. 예측 함수 정의
def predict_emotion_color(text):
    # 실제 서비스에서는 저장된 모델과 토크나이저를 불러와 사용해야 합니다.
    # 예시에서는 학습에 사용한 객체를 그대로 사용합니다.
    seq = tokenizer.texts_to_sequences([text])
    pad = pad_sequences(seq, maxlen=Config.MAX_LEN)
    pred = model.predict(pad)

    label_index = np.argmax(pred)
    label = le.inverse_transform([label_index])[0]
    prob = np.max(pred) * 100

    return label, f"{prob:.2f}%"

# ✅ 13. 새로운 문장으로 예측 테스트
print("\n===== 새로운 문장 예측 테스트 =====")
samples = [
    "오늘 너무 행복하고 날아갈 것 같아!",
    "진짜 화가 머리 끝까지 났어",
    "그냥 아무 생각 없어",
    "밤길을 혼자 걸으니 너무 무서웠어",
    "우와! 완전 서프라이즈잖아!",
    "그 사람 때문에 기분이 더러워졌어"
]

for s in samples:
    label, prob = predict_emotion_color(s)
    print(f"'{s}'\n    ➤ 예측 결과: {label} (신뢰도: {prob})")
```

```python
# ✅ 0. Config 클래스 및 라이브러리 임포트
import json
import os
import re
import urllib.request
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, models, optimizers, callbacks

# ----------------------------------------------------------------
# 하이퍼파라미터 및 주요 설정값을 관리하는 클래스
# ----------------------------------------------------------------
class Config:
    VOCAB_SIZE = 10000          # 단어 집합의 크기
    MAX_LEN_ENCODER = 32        # 인코더 입력의 최대 길이
    MAX_LEN_DECODER = 35        # 디코더 입출력의 최대 길이
    EMBEDDING_DIM = 256         # 임베딩 벡터의 차원
    LATENT_DIM = 512            # 인코더/디코더의 LSTM 유닛 수
    EPOCHS = 50                 # 최대 학습 횟수
    BATCH_SIZE = 64             # 배치 크기
    MODEL_SAVE_PATH = '/content/empathy_chatbot_model.weights.h5' # 파일 확장자 수정
    ENCODER_MODEL_SAVE_PATH = '/content/empathy_encoder_model.h5'
    DECODER_MODEL_SAVE_PATH = '/content/empathy_decoder_model.h5'
    TOKENIZER_SAVE_PATH = '/content/empathy_tokenizer.json'

# ✅ 1. 데이터 다운로드 및 불러오기
# 실습을 위해 '챗봇데이터'를 다운로드합니다. (출처: https://github.com/songys/Chatbot_data)
file_name = 'ChatbotData.csv'
file_path = f'/content/{file_name}'
url = 'https://github.com/songys/Chatbot_data/raw/master/ChatbotData.csv'

if not os.path.exists(file_path):
    print(f"'{file_name}'을 다운로드합니다...")
    urllib.request.urlretrieve(url, file_path)
    print("다운로드 완료!")
else:
    print(f"'{file_name}'이 이미 존재합니다.")

df = pd.read_csv(file_path)
# 질문(Q)을 '감정 문장', 답변(A)을 '위로 문장'으로 간주하고 사용합니다.
df = df[['Q', 'A']]
df.rename(columns={'Q': 'utterance', 'A': 'response'}, inplace=True)

# ✅ 2. 전처리
# 간단한 전처리 함수
def preprocess_sentence(sentence):
    sentence = sentence.lower().strip()
    sentence = re.sub(r"([?.!,])", r" \1 ", sentence)
    sentence = re.sub(r'[" "]+', " ", sentence)
    sentence = re.sub(r"[^a-zA-Z가-힣0-9?.!,]+", " ", sentence)
    sentence = sentence.strip()
    return sentence

df['utterance'] = df['utterance'].apply(preprocess_sentence)
df['response'] = df['response'].apply(preprocess_sentence)

# 생성 모델 학습을 위해 디코더의 입력과 출력 데이터 생성
# 출력 문장의 시작과 끝에 <sos>, <eos> 토큰 추가
df['decoder_input'] = df['response'].apply(lambda x : '<sos> ' + x)
df['decoder_output'] = df['response'].apply(lambda x : x + ' <eos>')

# ✅ 3. 토크나이징
# 질문과 답변의 모든 단어를 하나의 토크나이저에 학습
tokenizer = Tokenizer(num_words=Config.VOCAB_SIZE, filters='', oov_token='<unk>')
tokenizer.fit_on_texts(df['utterance'])
tokenizer.fit_on_texts(df['decoder_input'])
tokenizer.fit_on_texts(df['decoder_output'])

# 단어 -> 정수 변환
encoder_input = tokenizer.texts_to_sequences(df['utterance'])
decoder_input = tokenizer.texts_to_sequences(df['decoder_input'])
decoder_output = tokenizer.texts_to_sequences(df['decoder_output'])

# 패딩
encoder_input = pad_sequences(encoder_input, maxlen=Config.MAX_LEN_ENCODER, padding='post')
decoder_input = pad_sequences(decoder_input, maxlen=Config.MAX_LEN_DECODER, padding='post')
decoder_output = pad_sequences(decoder_output, maxlen=Config.MAX_LEN_DECODER, padding='post')

# ✅ 4. 데이터 분할
x_train, x_test, y_train_in, y_test_in, y_train_out, y_test_out = train_test_split(
    encoder_input, decoder_input, decoder_output, test_size=0.2, random_state=42
)

# ✅ 5. Seq2Seq 모델 정의 (학습용)
# --- 인코더 ---
encoder_inputs = layers.Input(shape=(None,), name='encoder_input')
encoder_embedding = layers.Embedding(Config.VOCAB_SIZE, Config.EMBEDDING_DIM, name='encoder_embedding')(encoder_inputs)
# return_state=True 로 마지막 시점의 은닉 상태(h)와 셀 상태(c)를 받음
_, state_h, state_c = layers.LSTM(Config.LATENT_DIM, return_state=True, name='encoder_lstm')(encoder_embedding)
encoder_states = [state_h, state_c] # 컨텍스트 벡터

# --- 디코더 ---
decoder_inputs = layers.Input(shape=(None,), name='decoder_input')
decoder_embedding_layer = layers.Embedding(Config.VOCAB_SIZE, Config.EMBEDDING_DIM, name='decoder_embedding')
decoder_embedding = decoder_embedding_layer(decoder_inputs)
# return_sequences=True 로 모든 시점의 출력을 받음
decoder_lstm = layers.LSTM(Config.LATENT_DIM, return_sequences=True, return_state=True, name='decoder_lstm')
# 디코더는 인코더의 마지막 상태를 초기 상태(initial_state)로 사용
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = layers.Dense(Config.VOCAB_SIZE, activation='softmax', name='decoder_dense')
decoder_outputs = decoder_dense(decoder_outputs)

# --- 모델 결합 ---
model = models.Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()


# ✅ 6. 모델 학습
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
model.fit(
    [x_train, y_train_in], y_train_out,
    validation_data=([x_test, y_test_in], y_test_out),
    epochs=Config.EPOCHS,
    batch_size=Config.BATCH_SIZE,
    callbacks=[early_stopping]
)


# ✅ 7. 예측(추론) 모델 생성
# 추론 시에는 한 단어씩 생성해야 하므로, 학습 모델과 별도의 모델을 만듦

# --- 인코더 모델 ---
encoder_model = models.Model(encoder_inputs, encoder_states)

# --- 디코더 모델 ---
decoder_state_input_h = layers.Input(shape=(Config.LATENT_DIM,))
decoder_state_input_c = layers.Input(shape=(Config.LATENT_DIM,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

decoder_embedding_2 = decoder_embedding_layer(decoder_inputs)
decoder_outputs_2, state_h_2, state_c_2 = decoder_lstm(decoder_embedding_2, initial_state=decoder_states_inputs)
decoder_states_2 = [state_h_2, state_c_2]
decoder_outputs_2 = decoder_dense(decoder_outputs_2)

decoder_model = models.Model(
    [decoder_inputs] + decoder_states_inputs,
    [decoder_outputs_2] + decoder_states_2
)

# ✅ 8. 토크나이저 및 모델 저장
# 토크나이저 저장
tokenizer_json = tokenizer.to_json()
with open(Config.TOKENIZER_SAVE_PATH, 'w', encoding='utf-8') as f:
    f.write(json.dumps(tokenizer_json, ensure_ascii=False))

# 모델 가중치 저장
model.save_weights(Config.MODEL_SAVE_PATH)
encoder_model.save(Config.ENCODER_MODEL_SAVE_PATH)
decoder_model.save(Config.DECODER_MODEL_SAVE_PATH)
print(f"✅ 모델 및 토크나이저 저장 완료")


# ✅ 9. 예측 함수 정의
# 정수 -> 단어 변환을 위한 사전
index_to_word = tokenizer.index_word

def generate_response(input_sentence):
    input_seq = tokenizer.texts_to_sequences([preprocess_sentence(input_sentence)])
    input_seq = pad_sequences(input_seq, maxlen=Config.MAX_LEN_ENCODER, padding='post')

    states_value = encoder_model.predict(input_seq, verbose=0)

    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = tokenizer.word_index['<sos>']

    stop_condition = False
    decoded_sentence = ''

    while not stop_condition:
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value, verbose=0)

        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = index_to_word.get(sampled_token_index, '')

        if sampled_word == '<eos>' or len(decoded_sentence.split()) > Config.MAX_LEN_DECODER:
            stop_condition = True
            continue

        decoded_sentence += ' ' + sampled_word

        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index
        states_value = [h, c]

    return decoded_sentence.strip()


# ✅ 10. 새로운 문장으로 예측 테스트
print("\n===== 새로운 문장 예측 테스트 =====")
samples = [
    "오늘 너무 행복하고 날아갈 것 같아!",
    "진짜 화가 머리 끝까지 났어",
    "그냥 아무 생각 없어",
    "밤길을 혼자 걸으니 너무 무서웠어",
    "그 사람 때문에 기분이 더러워졌어",
    "시험에 떨어져서 너무 우울해"
]

for s in samples:
    response = generate_response(s)
    print(f"입력: '{s}'")
    print(f"   ➤ AI의 위로: {response}\n")
```

```python
# ✅ 0. Config 클래스 및 라이브러리 임포트
import json
import os
import urllib.request
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer, tokenizer_from_json
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, models, callbacks

# ----------------------------------------------------------------
# 하이퍼파라미터 및 주요 설정값을 관리하는 클래스
# ----------------------------------------------------------------
class Config:
    VOCAB_SIZE = 10000
    MAX_LEN = 32
    EMBEDDING_DIM = 128
    EPOCHS = 10  # 추가 학습을 위한 에포크 수
    BATCH_SIZE = 64
    MODEL_SAVE_PATH = '/content/rgbcmyw_emotion_model.h5'
    TOKENIZER_SAVE_PATH = '/content/rgbcmyw_tokenizer.json'

    # (수정) 추가 학습용 데이터 URL 및 파일 경로를 유효한 것으로 변경
    NEW_DATA_URL = 'https://github.com/ohgzone/file1/raw/main/aihub_coupus.csv'
    NEW_DATA_FILENAME = 'aihub_coupus.csv' # 파일명도 일치시킴
    NEW_DATA_FILE_PATH = f'/content/{NEW_DATA_FILENAME}'

# ----------------------------------------------------------------
# (추가) 파일 다운로드 유틸리티 함수
# ----------------------------------------------------------------
def download_file(url, file_path, file_description):
    """지정된 URL에서 파일을 다운로드하는 함수"""
    if not os.path.exists(file_path):
        print(f"'{file_description}'({os.path.basename(file_path)})을 다운로드합니다...")
        try:
            urllib.request.urlretrieve(url, file_path)
            print("다운로드 완료!")
            return True
        except Exception as e:
            print(f"🚨 다운로드 실패: {e}")
            return False
    else:
        print(f"'{file_description}'({os.path.basename(file_path)})이 이미 존재합니다.")
        # 파일이 이미 존재하므로 성공으로 간주
        return True

# ================================================================
# ✅ 1. 기존 모델 및 토크나이저 불러오기
# ================================================================
print("===== 기존 모델 및 토크나이저 불러오기 =====")

# 기존 모델과 토크나이저가 있어야 추가 학습이 가능합니다.
if not os.path.exists(Config.MODEL_SAVE_PATH) or not os.path.exists(Config.TOKENIZER_SAVE_PATH):
    print("🚨 모델 또는 토크나이저 파일이 없습니다.")
    print("먼저 원본 학습 코드를 실행하여 'rgbcmyw_emotion_model.h5'와 'rgbcmyw_tokenizer.json' 파일을 생성해주세요.")
    # exit() 대신 return 또는 다른 흐름 제어를 사용하여 Colab 환경에서 셀 실행이 중단되지 않도록 함
    # 이 경우, 아래 코드가 실행되지 않도록 new_df를 None으로 유지
    new_df = None
else:
    # 저장된 모델 불러오기
    model = models.load_model(Config.MODEL_SAVE_PATH)
    print(f"✅ 모델 로드 완료: {Config.MODEL_SAVE_PATH}")
    model.summary()

    # 저장된 토크나이저 불러오기
    with open(Config.TOKENIZER_SAVE_PATH, 'r', encoding='utf-8') as f:
        tokenizer_json = json.load(f)
        tokenizer = tokenizer_from_json(tokenizer_json)
    print(f"✅ 토크나이저 로드 완료: {Config.TOKENIZER_SAVE_PATH}")


    # ================================================================
    # ✅ 2. 추가 학습용 데이터 다운로드 및 불러오기
    # ================================================================
    print("\n===== 추가 학습용 데이터 준비 =====")

    new_df = None
    if download_file(Config.NEW_DATA_URL, Config.NEW_DATA_FILE_PATH, "추가 학습 데이터"):
        try:
            new_df = pd.read_csv(Config.NEW_DATA_FILE_PATH)
            new_df.rename(columns={'문장': 'utterance', '감정': 'emotion'}, inplace=True)
            print("\n추가할 데이터 샘플:")
            print(new_df.head())
        except FileNotFoundError:
            print(f"🚨 추가 학습 데이터 파일({Config.NEW_DATA_FILENAME})을 찾을 수 없습니다.")
        except Exception as e:
            print(f"🚨 파일 로딩 중 오류 발생: {e}")
            new_df = None

# ✅ 3. 새로운 데이터 전처리 및 모델 추가 학습
# new_df가 성공적으로 로드된 경우에만 이 블록을 실행
if new_df is not None and 'model' in locals():
    print("\n===== 새로운 데이터 전처리 및 추가 학습 =====")

    emotion_map = {
        '분노': 'r', '슬픔': 'b', '기쁨': 'y', '행복': 'y', '불안': 'b',
        '공포': 'c', '혐오': 'm', '놀람': 'w', '지루함': 'g',
        '중립': 'g', 'neutral': 'g'
    }
    new_df['label'] = new_df['emotion'].map(emotion_map)
    new_df = new_df.dropna(subset=['label', 'utterance'])

    new_sequences = tokenizer.texts_to_sequences(new_df['utterance'])
    x_new = pad_sequences(new_sequences, maxlen=Config.MAX_LEN)

    # 라벨 인코더는 기존 학습과 일관성을 유지해야 함
    le = LabelEncoder()
    all_labels = sorted(list(set(emotion_map.values()))) # 정렬하여 일관성 유지
    le.fit(all_labels)
    y_new = le.transform(new_df['label'])

    print("\n모델 추가 학습을 시작합니다...")
    early_stopping_new = callbacks.EarlyStopping(monitor='loss', patience=2, restore_best_weights=True)

    # 모델을 다시 컴파일하여 학습을 준비
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    history_new = model.fit(
        x_new, y_new,
        epochs=Config.EPOCHS,
        batch_size=Config.BATCH_SIZE,
        callbacks=[early_stopping_new],
        validation_split=0.1 # 추가 데이터의 일부를 검증용으로 사용
    )

    print("\n✅ 추가 학습 완료!")

    # ✅ 4. 추가 학습 후 모델 저장
    model.save(Config.MODEL_SAVE_PATH)
    print(f"✅ 추가 학습된 모델 저장 완료: {Config.MODEL_SAVE_PATH}")


# ✅ 5. 예측 테스트 (추가 학습된 모델 사용)
# 'model' 변수가 로드되었는지 확인 후 진행
if 'model' in locals():
    def predict_emotion_color_updated(text, loaded_model, loaded_tokenizer, label_encoder):
        seq = loaded_tokenizer.texts_to_sequences([text])
        pad = pad_sequences(seq, maxlen=Config.MAX_LEN)
        pred = loaded_model.predict(pad, verbose=0)
        label_index = np.argmax(pred)
        # 라벨 인코더가 fit되어 있어야 inverse_transform 가능
        if hasattr(label_encoder, 'classes_'):
             label = label_encoder.inverse_transform([label_index])[0]
        else:
             # Fallback or error handling
             label = "N/A"
        prob = np.max(pred) * 100
        return label, f"{prob:.2f}%"

    print("\n===== 추가 학습 후 예측 테스트 =====")
    samples = [
        "오늘 너무 행복하고 날아갈 것 같아!",
        "진짜 화가 머리 끝까지 났어",
        "그냥 아무 생각 없어",
        "밤길을 혼자 걸으니 너무 무서웠어",
        "우와! 완전 서프라이즈잖아!",
        "그 사람 때문에 기분이 더러워졌어",
        "마감일이 다가오니 초조하고 불안하다",
        "강의가 너무 지루해서 졸았어"
    ]

    # 라벨 인코더 준비
    if 'le' not in locals():
        emotion_map_pred = {
            '분노': 'r', '슬픔': 'b', '기쁨': 'y', '행복': 'y', '불안': 'b',
            '공포': 'c', '혐오': 'm', '놀람': 'w', '지루함': 'g',
            '중립': 'g', 'neutral': 'g'
        }
        le = LabelEncoder()
        all_labels_pred = sorted(list(set(emotion_map_pred.values())))
        le.fit(all_labels_pred)

    for s in samples:
        label, prob = predict_emotion_color_updated(s, model, tokenizer, le)
        print(f"'{s}'\n    ➤ 예측 결과: {label} (신뢰도: {prob})")
else:
    print("\n🚨 모델이 로드되지 않아 예측을 건너뜁니다.")
```

```python
# -*- coding: utf-8 -*-
"""
h5 모델 및 토크나이저를 로드하여 감정 분석을 수행하는 최종 예측 코드
"""
# ✅ 1. 필수 라이브러리 임포트
import json
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder

# ✅ 2. 설정값 정의
class Config:
    """예측에 필요한 주요 설정값을 관리합니다."""
    MAX_LEN = 32  # 모델 학습 시 사용한 문장의 최대 길이
    MODEL_SAVE_PATH = '/content/rgbcmyw_emotion_model.h5'
    TOKENIZER_SAVE_PATH = '/content/rgbcmyw_tokenizer.json'

# ✅ 3. 핵심 기능: 모델, 토크나이저, 라벨 인코더 로딩
def load_prediction_assets():
    """
    예측에 필요한 모델, 토크나이저, 라벨 인코더를 불러옵니다.
    Returns:
        tuple: (로드된 모델, 토크나이저, 라벨 인코더)
    """
    # 모델 로드
    try:
        model = tf.keras.models.load_model(Config.MODEL_SAVE_PATH)
        print(f"✅ 모델 로드 성공: {Config.MODEL_SAVE_PATH}")
    except Exception as e:
        print(f"🔥 모델 로드 실패: {e}")
        return None, None, None

    # 토크나이저 로드
    try:
        with open(Config.TOKENIZER_SAVE_PATH, 'r', encoding='utf-8') as f:
            tokenizer_json = json.load(f)
            tokenizer = tf.keras.preprocessing.text.tokenizer_from_json(tokenizer_json)
        print(f"✅ 토크나이저 로드 성공: {Config.TOKENIZER_SAVE_PATH}")
    except Exception as e:
        print(f"🔥 토크나이저 로드 실패: {e}")
        return None, None, None

    # 라벨 인코더 재생성 (학습 시와 동일한 순서 보장)
    # 모델의 숫자 출력을 실제 라벨(r, g, b 등)로 변환하기 위해 필수
    emotion_map = {
        '분노': 'r', '슬픔': 'b', '기쁨': 'y', '행복': 'y',
        '공포': 'c', '혐오': 'm', '놀람': 'w', '중립': 'g'
    }
    # sorted()를 통해 항상 일관된 순서의 라벨 인코더를 생성
    all_labels = sorted(list(set(emotion_map.values())))
    le = LabelEncoder()
    le.fit(all_labels)
    print(f"✅ 라벨 인코더 준비 완료. 클래스: {le.classes_}")

    return model, tokenizer, le

# ✅ 4. 예측 함수
def predict_emotion(text, model, tokenizer, le):
    """
    새로운 텍스트의 감정을 예측하여 색상 라벨과 신뢰도를 반환합니다.

    Args:
        text (str): 감정을 분석할 문장
        model (tf.keras.Model): 로드된 Keras 모델
        tokenizer (tf.keras.preprocessing.text.Tokenizer): 로드된 토크나이저
        le (LabelEncoder): 준비된 라벨 인코더

    Returns:
        tuple: (예측된 라벨, 신뢰도)
    """
    # 1. 텍스트 전처리: 시퀀스 변환 및 패딩
    seq = tokenizer.texts_to_sequences([text])
    padded_seq = pad_sequences(seq, maxlen=Config.MAX_LEN)

    # 2. 모델 예측
    prediction = model.predict(padded_seq, verbose=0)

    # 3. 결과 해석
    label_index = np.argmax(prediction)
    confidence = np.max(prediction) * 100
    predicted_label = le.inverse_transform([label_index])[0]

    return predicted_label, f"{confidence:.2f}%"


# ✅ 5. 메인 실행 블록
if __name__ == "__main__":
    # 1. 예측에 필요한 자원(모델, 토크나이저 등) 로드
    model, tokenizer, le = load_prediction_assets()

    # 2. 자원이 성공적으로 로드되었는지 확인
    if model and tokenizer and le:
        print("\n" + "="*50)
        print("          새로운 문장 감정 예측 테스트")
        print("="*50)

        # 3. 예측을 원하는 문장들
        sentences_to_predict = [
            "오늘 너무 행복하고 날아갈 것 같아!",
            "진짜 화가 머리 끝까지 났어.",
            "그냥 아무 생각 없어.",
            "밤길을 혼자 걸으니 너무 무서웠어.",
            "우와! 완전 서프라이즈잖아!",
            "그 사람 때문에 기분이 더러워졌어.",
            "이 영화 정말 감동적이고 슬펐어.",
            "너 때문에 내 계획이 다 망가졌잖아!",
            "로또에 당첨되다니 믿기지가 않아!",
            "그럭저럭 지낼만 해."
        ]

        # 4. 각 문장에 대해 예측 수행 및 결과 출력
        for sentence in sentences_to_predict:
            label, prob = predict_emotion(sentence, model, tokenizer, le)
            print(f"'{sentence}'\n    ➤ 예측 결과: {label} (신뢰도: {prob})")
            print("-" * 20)

    else:
        print("\n🔥 예측을 시작할 수 없습니다. 파일 경로 및 상태를 확인해주세요.")
```
